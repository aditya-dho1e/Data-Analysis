```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')  # optional: for ggplot-like style
```

```python
df_canada = pd.read_csv('d:/datasets/canadian_immegration_data.csv')
df_canada.head()
```

# Area Plot

```python
df_canada.set_index('Country', inplace=True)
df_canada.head()
```

```python
df_sorted = df_canada.sort_values(by='Total', ascending=False)
df_sorted.head()
```

```python
years = list(map(str, range(1980, 2014)))

top_five = df_sorted.loc[:'Pakistan', years].transpose()
df_sorted.head()
```

```python
top_five.plot(kind='area',
             stacked=False,
             figsize=(20, 10))
```

# Pie Charts

```python
continents = df_canada.groupby('Continent').sum()
continents
```

```python
continents['Total'].plot(kind='pie')

plt.title('Continent wise immegration')

plt.show()
```

```python
continents['Percentage'] = continents['Total']/continents['Total'].sum()
```

```python
explode = [0.1 if per < 0.1 else 0 for per in continents['Percentage']]

continents['Percentage'].plot(kind='pie', explode=explode, autopct='%1.1f%%')

plt.title('Continent wise immegration in Percentage')
plt.show()
```

# Scatter Plot

```python
df_canada.head()
```

```python
def add_row(row):
    ele = []
    for x in row:
        if isinstance(x, (int, float)):
            ele.append(x)
    return sum(ele)

tot = df_canada.loc[:, years].transpose().apply(add_row, axis=1)
tot.head()
```

```python
tot = pd.DataFrame(tot)
tot = tot.reset_index()
tot.columns = ['Year', 'Total']
```

```python
tot.head()
```

```python
tot.plot(kind='scatter',
        x='Year',
        y='Total')

plt.xticks(rotation='vertical')
plt.show()
```

**Bubble Plot**

```python
brazil = df_canada.loc[:, years].transpose()['Brazil']
```

```python
brazil_norm = (brazil - brazil.min()) / (brazil.max() - brazil.min())
brazil_norm
```

```python
tot.plot(kind='scatter',
         figsize=(14,8),
         alpha=0.5,
        x='Year',
        y='Total',
        s=brazil_norm * 2000 + 10)

plt.xticks(rotation='vertical')
plt.show()
```

# Plotting directly with matplotlib

```python
years = range(1980, 2014)
total = np.random.randint(0, 20000, (34,))
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(8,6))

ax[0].plot(years,total)
ax[0].set_title('Line')

ax[1].scatter(x = years, y = total)
ax[1].set_title('Scatter')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, sharey=True) # paras:- nrows, ncols, indices

ax[0].plot(years, total, 'g-')
ax[1].scatter(x=years, y=total, color='darkblue', marker='o')
```
